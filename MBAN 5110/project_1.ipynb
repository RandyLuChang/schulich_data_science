```python
import statsmodels.api as sm
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/changlu/Downloads/dataset_lm.csv')
```

```python
df.head()
```

***Part 1***

```python

X = df.drop("Dependent Var", axis=1)  
y = df["Dependent Var"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

summary = model.summary()
summary
```

```python
import matplotlib.pyplot as plt

num_vars = df.shape[1] - 1

fig, axes = plt.subplots(nrows=num_vars, figsize=(10, 4 * num_vars))

for i, col in enumerate(df.columns[1:]):
    axes[i].scatter(df[col], df["Dependent Var"], color='blue', alpha=0.5)
    axes[i].set_title(f'Dependent Var vs {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Dependent Var')

plt.tight_layout()
plt.show()
```

1. Based on the data set and model we built, the R-squared is 1, which implies that the model can explain 100% of the variance in the dependent variable. However, this is almost impossible in real world. This might indicate  overfitting or the data might be generated on purpose.

2. The coefficent constant is 32, which means that if all explanatory variables are set to zero, the expected value of dependent will be 32. 

    For each one unit of increase in Explantory Var #1, the dependent variable is expected to increase by 1.3 units.

    For each one unit of increase in Explantory Var #2, the dependent variable is expected to increase by 1.7 units.

    For each one unit of increase in Explantory Var #3, the dependent variable is expected to increase by 6.2 units.

    For each one unit of increase in Explantory Var #4, the dependent variable is expected to increase by 2.1 units.

    For each one unit of increase in Explantory Var #5, the dependent variable is expected to decrease by 0.9 units.

    Because other variables' coeeficient are way too small, they may be negligible or have very little effect on the dependent variable.

3. As for the p-values, explanatory var #1-7, 9, 10, and 11 have p-values that are close to 0, indicating they are statistically significant. For other explanatory variables, the p-values that are greater than 0.05 indicated that they may not be statistically significant predictors.

4. The Durbin-Watson statistic value is close to 2, which indicated that there's no autocorrelation in the residuals.
